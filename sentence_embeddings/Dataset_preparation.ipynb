# Data Preparation

```python
# sentences_list = []
i = 5000
for findex in trange(len(csv_file_list)):
  df = pd.read_csv(os.path.join(case_sentence_csv_folder, csv_file_list[findex]))
  encoded_sentences_list = []
  i+=1000
  for index, row in df.iterrows():
    sentence = row['sentence'].strip().lower()
    encode_sentence = model.encode(sentence)
    encoded_sentences_list.append(encode_sentence)
    # sentences_list.append(sentence)
  np_array = np.array(encoded_sentences_list)
  path = '/content/drive/Shareddrives/SigmaLaw-WPP/EAMS/datasets/encoded_sentences_768_{}.txt'.format(i)
  with open(path, 'w') as f:
    np.savetxt(f, np_array)
  print(path)
```
